```python
traceExport = pd.read_csv('ordersExport (4).csv', delimiter=',')
```

```python
display(traceExport)
```

```python
lineItems = pd.DataFrame(columns=['id', 'name', 'customer', 'shopifyId', 'total', 'createdAt', 'updatedAt', 'consignerId', 'storeId','inventories'])
```

```python
for i in range(len(traceExport)):
    bigstr = traceExport.iloc[i, 10]
    substr = '{"id":\d+,"code"' # finds starting of new entry
    x = re.findall(substr, bigstr)
    loopCount = len(x)
    #print(loopCount)
    
    # put stubs here
    nameStub = traceExport.iloc[i, 1]
    creatStub = traceExport.iloc[i, 6]
    updateStub = traceExport.iloc[i, 7]
    consignStub = traceExport.iloc[i, 3]
    #print(nameStub)
    
    for x in range(loopCount):
        
        # here we get the id
        groupItems = traceExport.iloc[i, 10]
        idstr = '"orderId":\d+,'
        idarr = re.findall(idstr, groupItems)
        #print(idarr)
        idFin = idarr[x][10:16]
        #print(idFin)
        
        # here we get the name
        nameFin = nameStub
        
        
        
        # here we get the shopifyId
        groupItems = traceExport.iloc[i, 10]
        sidstr = '"shopifyId":"\d+",'
        sidarr = re.findall(sidstr, groupItems)
        #print(sidarr)
        try:
            sidFin = sidarr[x][13:27]
        except:
            sidFin = 'null'
        #print(sidFin)
        
        # here we get the total
        groupItems = traceExport.iloc[i, 10]
        totstr = '"price":"\d+"'
        totarr = re.findall(totstr, groupItems)
        #print(totarr)
        totFin = totarr[x][9:13]
        #print(totFin)
        
        # here we get the created at
        creFin = creatStub
        
        # here we get the update at
        updFin = updateStub
        
        # here we get the consignerId
        groupItems = traceExport.iloc[i, 10]
        cidstr = '"consignerId":\d+,'
        cidarr = re.findall(cidstr, groupItems)
        #print(cidarr)
        cidFin = cidarr[x][14:18]
        #print(cidFin)
        
        
        
        
        
        #here we get the customer
        groupItems = traceExport.iloc[i, 10]
        constr = '"consignerId":\d+,'
        conarr = re.findall(constr, groupItems)
        #print(conarr)
        idFin = idarr[x][10:16]
        #print(idFin)
        
        
        
        
        
        
        #TODO
        
        # here we get the storeID
        groupItems = traceExport.iloc[i, 10]
        stostr = '"storeId":\d+,'
        stoarr = re.findall(stostr, groupItems)
        #print(stoarr)
        stoFin = stoarr[x][10:12]
        #print(stoFin)
        
        # here we get the inventory
        groupItems = traceExport.iloc[i, 10]
        invItems = re.split("},{", groupItems)
        #print(len(invItems))
        invFin = invItems[x]
        
        string = invFin
        pattern = '"code":"'
        pattern2 = '"option'
        match1=(re.search(pattern, string))
        #print(match1)
        match2=(re.search(pattern2, string))
        #print(match2)
        
        invFin = invFin[match1.start() + 8 :match2.start() - 1]
        
        entry = [idFin, nameFin, sidFin, totFin, creFin, updFin, cidFin, stoFin, invFin]
        
        temporary_df = pd.DataFrame([entry], columns=['id', 'name', 'shopifyId', 'total', 'createdAt', 'updatedAt', 'consignerId', 'storeId','inventories'])
        #print(temporary_df)
        
        #lineItems = lineItems.append(temporary_df, ignore_index=True)
        
        frames = [lineItems, temporary_df]
  
        lineItems = pd.concat(frames)
#print(lineItems)
        
```
